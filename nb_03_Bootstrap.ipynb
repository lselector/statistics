## Bootstrap - testing

Simple demo showing how bootstrap sampling works

```python
import sys, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
# import our utilities
from util_jupyter import *
from util_models import *
```

```python
N_init = 100000
np.random.seed()
# init_sample = np.random.normal(loc=2.5,scale=1.0, size=N_init)
init_sample = np.random.choice([2,3], size=N_init)
# init_sample = np.random.uniform(low=2.0, high=3.0, size=N_init)
# init_sample = np.random.triangular(left=2.0, mode=2.0, 
#                                    right=3.0, size=N_init)
# init_sample = np.random.exponential(scale=1.0, size=N_init)

print(init_sample[:20])
print(commify(len(init_sample)))
```

```python
# let's generate samples with replacement
N_samples=1000
size=1000
mysum=np.zeros(size) # array of means
for ii in range(N_samples):
    aa = np.random.choice(init_sample,size=size)
    mysum += aa
# mysum /= N_samples
```

```python
plt.rcParams["figure.figsize"] = (12, 4) # (width, height)
fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3)
ax1.hist(init_sample,bins=100)
ax2.hist(mysum,bins=50)
ax3.hist(aa,bins=50)
ax1.set_xlim(left=0,right=5)
# ax2.set_xlim(left=0,right=5)
ax3.set_xlim(left=0,right=5)
plt.show();
```
