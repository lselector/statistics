## Extracting signal from noise

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def make_noise(Npoints):
    return 4*(np.random.rand(Npoints) - 0.5)
```

```python
Npoints=201
```

```python
# clean sin signal
x = np.linspace(-np.pi, np.pi, Npoints)
y0 = np.sin(x)
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, y0 + make_noise(Npoints))
plt.show()
```

### Extract signal from noise by summation

```python
# Extract signal from noise
# Generate big number of noisy records
# Calculate sum of them
# Signal will grow ~ N
# Noise will grow ~ sqrt(N)
# S/N will grow ~ sqrt(N)

N=5000
y = np.zeros(Npoints)
for ii in range(N):
    y = y + y0 + make_noise(Npoints)
y = y/N
plt.plot(x, y)
plt.show()
```
