```python
import sys, os
from IPython.display import clear_output
import scipy.stats, numpy as np, pandas as pd, time, datetime, math
import matplotlib.pyplot as plt
# High-resolution plot outputs for Retina notebooks
%config InlineBackend.figure_format = 'retina'
```

```python
arr = list(range(101)) # 0,1,2, ...,99,100
arr = [0.01*x for x in arr] # 0, 0.01, 0.02, ..., 0.99, 1.0
xx = [x for x in arr]
y1mse = [x**2 for x in xx]
y2mse = [(1-x)**2 for x in xx]

x1 = xx[1:]
x2 = xx[:100]
#print(xx)
#print(x1)
#print(x2)
y1log = [-math.log(x) for x in x1]
y2log = [-math.log(1-x) for x in x2]
```

```python
plt.plot(xx,y1mse,color="red")
plt.plot(xx,y2mse,color="red")
plt.plot(x1,y1log,color="green")
plt.plot(x2,y2log,color="green")
plt.title("MSE (red),  LOG-LOSS (green)")
plt.grid(True)
plt.show(); # semicolon needed to remove text output on top
```
